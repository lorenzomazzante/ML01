# 1. Tutorial on a synthetic dataset
The objective of this first Section is to try to reproduce the results presented in the Figure of slide 34.

Compile the following cell to generate a synthetic dataset.

```python
# RPONSE  LA QUESTION 3.1 :
np.random.seed(407)

from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

# Generate the data
n_circles = 1000
X_circles, y_circles = make_circles(n_samples=n_circles, factor=.6, noise=.05, random_state=407)
Ycirlces_array = np.zeros((1000,1))
Ycirlces_array[:,0] = y_circles

n1 = 100
n2 = 100
n3 = 80
n4 = 50
n5 = 50
n6 = 50
n7 = 50
n = n_circles+n1+n2+n3+n4+n5+n6+n7
mu1 = [0, 0]
mu2 = [0, 4]
mu3 = [0,7]
mu4 = [-5,-1]
mu5 = [-5,2]
mu6 = [5,0]
mu7 = [5,-3]
Cov1 = np.array([[2, 1], [1, 3]])
Cov2 = np.array([[6, -3], [-3, 3.5]])
Cov3 = np.array([[1, 0], [0, 1]])
Cov4 = np.array([[1, 0], [0, 1]])
Cov5 = np.array([[1, 0], [0, 1]])
Cov6 = np.array([[1, 0], [0, 1]])
Cov7 = np.array([[1, 0], [0, 1]])
rng = np.random.default_rng(407)
XTrain = np.r_[X_circles*20, rng.multivariate_normal(mu1, Cov1, size=n1), rng.multivariate_normal(mu2, Cov2, size=n2), rng.multivariate_normal(mu3, Cov3, size=n3), rng.multivariate_normal(mu4, Cov4, size=n4), rng.multivariate_normal(mu5, Cov5, size=n5), rng.multivariate_normal(mu6, Cov6, size=n6), rng.multivariate_normal(mu7, Cov7, size=n7)]
YTrain = np.r_[Ycirlces_array, np.zeros((n1,1)), np.ones((n2,1)), np.zeros((n3,1)), np.ones((n4,1)), np.zeros((n5,1)), np.zeros((n6,1)), np.ones((n7,1))].ravel()


figScatter = plt.figure(figsize=(9,9))
ax1 = figScatter.add_subplot(1,1,1)
ax1.scatter(XTrain[np.where(YTrain==0),0], XTrain[np.where(YTrain==0),1], color='purple', marker='.', label='C0',s=140)
ax1.scatter(XTrain[np.where(YTrain==1),0], XTrain[np.where(YTrain==1),1], color='orange', marker='.', label='C1',s=140)
ax1.legend(fontsize=20, loc='upper right')
ax1.set_xlabel("$X_1$",fontsize=22)
ax1.set_ylabel("$X_2$",fontsize=22)
ax1.set_title("Scatter plot of the Training set",fontsize=24)
ax1.tick_params(axis='x', labelsize=22)
ax1.tick_params(axis='y', labelsize=22)

figScatter.savefig("figScatter_DL_Training_2D.pdf")
```
